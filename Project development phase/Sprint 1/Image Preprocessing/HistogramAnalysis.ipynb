```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
```

```python
I = imread('/content/9a06094f-8af7-49ec-a191-0f8c663d82f2___Rut._Bact.S 0794.JPG')
J = imread('/content/9a078cfa-5766-47c5-9bc0-021ba1495045___Rut._Bact.S 0790.JPG')
```

```python
plt.figure()
plt.subplot(121), plt.imshow(I)
plt.subplot(122), plt.imshow(J)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.abs(I[:, :, 0].astype(float) - J[:, :, 0].astype(float)), cmap='gray')
plt.show()
```
